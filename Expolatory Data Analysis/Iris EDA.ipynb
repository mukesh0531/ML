# EDA - Exploratory Data Analysis

EDA is an approach to understand the data - Mostly we use statistical graphics to understand but we do have non graphical analysis through which we can perform data analysis. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_data = pd.read_csv('Iris.csv')
#iris_data['Species'] = iris_data['Species'].strip()
iris_setosa = iris_data[iris_data['Species']=="setosa"]
iris_versicolor = iris_data[iris_data['Species']=='versicolor']
iris_virginica = iris_data[iris_data['Species']=='virginica']

print('# of data points for setosa:', iris_setosa.shape)
print('# of data points for versicolor:',iris_versicolor.shape)
print('# of data points for virginica:',iris_virginica.shape)
#iris_data.head()

iris_data.count()
iris_data['Species'].value_counts()
```

```python
# explore iris_data

print("List of Iris Variables: \n", iris_data.columns)
iris_data.tail()
```

```python
### Univariate analysis
plt.figure()
plt.subplot(121)
plt.scatter(iris_setosa['Sepal length'],np.zeros(50))
plt.scatter(iris_versicolor['Sepal length'],np.zeros(50))
plt.scatter(iris_virginica['Sepal length'],np.zeros(49))
plt.legend()
plt.subplot(122)
plt.scatter(iris_setosa['Sepal width'],np.zeros(50))
plt.scatter(iris_versicolor['Sepal width'],np.zeros(50))
plt.scatter(iris_virginica['Sepal width'],np.zeros(49))
plt.legend()
plt.show()

plt.figure()
plt.subplot(121)
plt.scatter(iris_setosa['Petal length'],np.zeros(50))
plt.scatter(iris_versicolor['Petal length'],np.zeros(50))
plt.scatter(iris_virginica['Petal length'],np.zeros(49))
plt.legend()
plt.subplot(122)
plt.scatter(iris_setosa['Petal width'],np.zeros(50))
plt.scatter(iris_versicolor['Petal width'],np.zeros(50))
plt.scatter(iris_virginica['Petal width'],np.zeros(49))
plt.legend()
plt.show()
```

1. Sepal Length is having overlapping points.
2. Sepal width some points are overlappint
3. PL and PW - we can clarly distinguish the one kind of flowers...

```python
sns.distplot(iris_data['Sepal length'],rug='true')
plt.show()
```

```python
#plt.figure()
#plt.subplot(221)
sns.FacetGrid(iris_data,hue='Species',size=5).map(sns.distplot,'Sepal length')
#plt.subplot(222)
sns.FacetGrid(iris_data,hue='Species',size=5).map(sns.distplot,'Sepal width')
#plt.subplot(223)
sns.FacetGrid(iris_data,hue='Species',size=5).map(sns.distplot,'Petal length')
#plt.subplot(224)
sns.FacetGrid(iris_data,hue='Species',size=5).map(sns.distplot,'Petal width')
plt.show()
```

```python
sns.FacetGrid(iris_data,col='Species',size=5).map(plt.hist,'Sepal length',bins=5)
plt.show()
```

```python
sns.boxplot(x='Species',y='Sepal length',data=iris_data)
plt.legend()
plt.show()
```

```python
sns.pairplot(data=iris_data,hue='Species',size=4)
plt.show()
```

```python
sns.jointplot(x="Petal length", y="Petal width", data=iris_setosa, kind="kde",size=10);
plt.show();
```
